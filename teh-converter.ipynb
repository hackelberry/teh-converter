```python
from mido import MidiFile
```

```python
mid = MidiFile('test.mid')
```

```python
tempo = 500000

notes = []

for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    
    dt = 0
    
    for msg in track:
        dt += msg.time
        
        if msg.type == 'set_tempo':
            tempo = msg.tempo
               
        if msg.type == 'note_on':
            notes.append((msg.note, dt * tempo / (96 * 1000)))
```

```python
import operator
```

```python
notes = list(set(notes))
notes.sort(key=operator.itemgetter(1))
```

```python
strategy = ' basic-high'

def convert(note):
    # uses the playable midi notes between 52 and 76 (E3 and E5)
    # and ignores everything else
    if strategy == 'basic-high':
        if note == 52:
            return 0
        if note == 57:
            return 1
        if note == 62:
            return 2
        if note == 67:
            return 3
        if note == 71:
            return 4
        if note == 76:
            return 5
        
    # assumes midi notes 0-5 correspond to strings 0-5
    # and ignores everything else
    if strategy == 'basic':
        if note == 0:
            return 0
        if note == 1:
            return 1
        if note == 2:
            return 2
        if note == 3:
            return 3
        if note == 4:
            return 4
        if note == 5:
            return 5
        
    # transposes all playable notes to the playable octave and ignores everything else.
    # does not use the high E (6th) string
    if strategy == 'ignore':
        if note % 12 == 4:
            return 0
        if note % 12 == 9:
            return 1
        if note % 12 == 2:
            return 2
        if note % 12 == 7:
            return 3
        if note % 12 == 11:
            return 4
        
    # uses a reasonably close playable note, probably sounds like garbage.
    # does not use the high E (6th) string
    if strategy == 'closest':
        if note % 12 in [4, 5, 6]:
            return 0
        if note % 12 == [9, 10]:
            return 1
        if note % 12 == [2, 3]:
            return 2
        if note % 12 == [7, 8]:
            return 3
        if note % 12 == [11, 0, 1]:
            return 4
```

```python
for i in range(len(notes)):
    print('hits[', i, '].string = ', convert(notes[i][0]), ';\n',
          'hits[', i, '].time = ', int(notes[i][1]), ';\n', sep='')
```
